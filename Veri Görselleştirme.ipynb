```python
df['Date'].dtype # tarih olmasına rağmen veri türü Object
```

```python
# Veri türünü değiştir
df['Date']=pd.to_datetime(df['Date'])
df['Date'].info()
```

```python
#Günü ayır
df['day']=df.Date.dt.day # Date sütünundaki günü al ve day sütunuoluşturup ekle
df['day']
```

```python
#Ay ı ayır
df['month']=df.Date.dt.month
df['month']
```

```python
df['year']=df.Date.dt.year
df['year']
```

```python
df[['day','month','year']]
```

```python
df['dayofweek']=df.Date.dt.dayofweek
df['dayofweek']
```

```python
df['dayofweek'].unique()
```

```python
# haftanın kaçıncı günü olduğu yerine gün adı yazsn
df['dayofweek'].replace({0:'Monday',1:'Tuesday',
                         2:'Wednesday',3:'Thursday',4:'Friday',
                         5:'Saturday',6:'Sunday'},inplace=True)
```

```python
df['dayofweek']
```

```python
#İkinci yöntem
df['day_name']=df.Date.dt.day_name()
df['day_name']
```

```python
df['month_name']=df.Date.dt.month_name()
df['month_name']
```

```python
df['Time'],df['Time'].dtype
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Time'].info()
```

```python
df['hour']=df.Time.dt.hour
df['hour']
```

```python
plt.figure(figsize=(12,6)) # grafiğin boyutları
sns.countplot(x=df['month_name']) 
# belirlenen kategorilerdeki toplamları alır
#Aylardaki alışveriş sayısı
plt.title('Aylardaki alışveriş sayısı')
plt.xticks(rotation=90) # etiket yazısını açısı
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Şueberin aylara göre alışveriş sayısı')
sns.countplot(x=df['month_name'],
              hue=df['Branch'])
```

```python
plt.title('Cinsiyete göre alışveriş sayısı')
sns.countplot(x=df['Gender'])
```

```python
plt.title('Cinsiyete göre aylık alış veriiş sayıları')
sns.countplot(x=df['month_name'], hue=df['Gender'])
```

```python
sns.countplot(y=df['month_name'], hue=df['Gender'])
```

```python
df['month_name'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
df['Product line'].value_counts().plot.pie(autopct='%1.1f%%',
                                           ylabel='',
                                           xlabel='',
                                           title='Grafik başlığı',
                                           explode=[.1,.1,.5,.1,.1,.1])               
```

```python
#sns pie graphic
sns.set(style='whitegrid')
plt.pie(df['Product line'].value_counts(),
       autopct='%1.1f%%',
        labels=df['Product line'].value_counts().index,
       startangle=90);
```

```python
sns.countplot(x=df['Product line'],hue=df['Gender'])
plt.xticks(rotation=30);
```
